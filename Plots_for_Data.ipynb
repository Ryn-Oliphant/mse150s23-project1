```python
filename = 'raw_data/Sp22_245L_sec-001_tensiletest-pps_bulk.raw' #sys.argv[1]

data = np.loadtxt(fname=filename, delimiter=' ', comments='"', usecols=range(6))
```

```python
def youngs_modulus (data):
    
    lin_fit = np.polyfit(data[0:200,5],data[0:200,1],1)
    lin_func = np.poly1d(lin_fit)
    
    print("Young's Modulus = ",lin_fit[0],'Pa')
```
